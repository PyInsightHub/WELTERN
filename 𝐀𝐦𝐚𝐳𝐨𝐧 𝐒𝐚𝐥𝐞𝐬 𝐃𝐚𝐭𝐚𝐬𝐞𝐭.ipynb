### *Feature Engineering and Visualization*

```python
# Extract temporal features
df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
df['DayOfWeek'] = df['OrderDate'].dt.day_name()

# Create 'DiscountAmount' and 'NetPrice'
# Discount is likely a percentage (e.g., 0.1 for 10%)
df['DiscountAmount'] = df['Discount'] * df['UnitPrice'] * df['Quantity']
df['NetPrice'] = df['UnitPrice'] * (1 - df['Discount'])

print("Features created: Year, Month, DayOfWeek, DiscountAmount, NetPrice")
display(df[['OrderDate', 'Year', 'Month', 'DayOfWeek', 'DiscountAmount', 'NetPrice']].head())
```

```python
# 1. Aggregate total sales per month
monthly_sales = df.groupby('Month')['TotalAmount'].sum().reset_index()

# 2. Visualize Monthly Sales Trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month', y='TotalAmount', marker='o', color='teal')
plt.title('Total Sales Trend per Month')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(range(1, 13))
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 3. Compare average DiscountAmount across Category
avg_discount_category = df.groupby('Category')['DiscountAmount'].mean().sort_values(ascending=False).reset_index()

# 4. 2D Bar Plot for average DiscountAmount per Category
plt.figure(figsize=(14, 7))
sns.barplot(data=avg_discount_category, x='Category', y='DiscountAmount', hue='Category', palette='magma', legend=False)
plt.title('Average Discount Amount per Category')
plt.xlabel('Category')
plt.ylabel('Mean Discount Amount')
plt.xticks(rotation=45)
plt.show()

# 5. Re-sampling to include new features DiscountAmount, Quantity, and TotalAmount

df_sample_new = df.sample(n=5000, random_state=42)
fig = px.scatter_3d(df_sample_new, x='DiscountAmount', y='Quantity', z='TotalAmount',
                    color='Category', opacity=0.7,
                    title='3D Interaction: DiscountAmount, Quantity, and TotalAmount')
fig.update_layout(margin=dict(l=0, r=0, b=0, t=40))
fig.show()
```
