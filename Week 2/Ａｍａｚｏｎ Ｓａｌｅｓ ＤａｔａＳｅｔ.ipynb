```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### **Ôº¨ÔΩèÔΩÅÔΩÑ Ôº¥ÔΩàÔΩÖ Ôº§ÔΩÅÔΩîÔΩÅÔΩìÔΩÖÔΩî ÔºÜ Ôº∞ÔΩíÔΩÖÔΩñÔΩâÔΩÖÔΩó Ôº¥ÔΩàÔΩÖ Ôº¶ÔΩâÔΩíÔΩìÔΩî Ôº¶ÔΩÖÔΩó Ôº≤ÔΩèÔΩóÔΩì**

```python
df = pd.read_csv('D:/ùêèùêòùêìùêáùêéùêç ùüê/ùüêùüéùüêùüî/ùêâùêöùêßùêÆùêöùê´ùê≤/Amazon Sales Dataset/Amazon.csv')
print('--- First 5 rows ---')
display(df.head())
```

### **Ôº§ÔΩâÔΩìÔΩêÔΩåÔΩÅÔΩô Ôº≥ÔΩïÔΩçÔΩçÔΩÅÔΩíÔΩô Ôº©ÔΩéÔΩÜÔΩèÔΩíÔΩçÔΩÅÔΩîÔΩâÔΩèÔΩé**

```python
print("\n--- DataFrame Info ---")
df.info()
```

### **Ôº∞ÔΩíÔΩÖÔΩñÔΩâÔΩÖÔΩó Ôº¥ÔΩàÔΩÖ Ôº¨ÔΩÅÔΩìÔΩî Ôº¶ÔΩÖÔΩó Ôº≤ÔΩèÔΩóÔΩì**

```python
print('--- Last 5 rows ---')
display(df.tail())
```

```python
info_df = pd.DataFrame({
    "Null %": (df.isnull().mean() * 100).round(2)
})

display(info_df)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='Blues')
plt.title('Missing Values')
plt.show()
```

### **Ôº£ÔΩèÔΩéÔΩñÔΩÖÔΩíÔΩî ÔºØÔΩíÔΩÑÔΩÖÔΩíÔº§ÔΩÅÔΩîÔΩÖ Ôº¥ÔΩè Ôº§ÔΩÅÔΩîÔΩÖÔΩîÔΩâÔΩçÔΩÖ ÔºØÔΩÇÔΩäÔΩÖÔΩÉÔΩîÔΩì**

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
duplicate_count = df.duplicated().sum()
summary_df = pd.DataFrame({
    "Metric": ["Total Rows (Before)", "Duplicate Rows Found"],
    "Value": [len(df), duplicate_count]
})
display(summary_df)
if duplicate_count > 0:
    df = df.drop_duplicates()
final_info = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.astype(str),
    "Non-Null Count": df.notnull().sum(),
    "Null Count": df.isnull().sum()
})
display(final_info)
print("‚úÖ Data Cleaning Completed Successfully.")
```

### **Ôº°ÔΩéÔΩÅÔΩåÔΩôÔΩöÔΩÖ Ôº≠ÔΩèÔΩéÔΩîÔΩàÔΩåÔΩô Ôº≥ÔΩÅÔΩåÔΩÖÔΩì Ôº¥ÔΩíÔΩÖÔΩéÔΩÑÔΩì**

```python
df['YearMonth'] = df['OrderDate'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('YearMonth')['TotalAmount'].sum().reset_index()
```

```python
monthly_sales = monthly_sales.sort_values('YearMonth')
print("--- Monthly Sales Summary ---")
print(monthly_sales.head())
```

### **Ôº©ÔΩÑÔΩÖÔΩéÔΩîÔΩâÔΩÜÔΩô Ôº¢ÔΩÖÔΩìÔΩîÔºçÔº≥ÔΩÖÔΩåÔΩåÔΩâÔΩéÔΩá Ôº∞ÔΩíÔΩèÔΩÑÔΩïÔΩÉÔΩîÔΩì**

```python
product_sales = df.groupby('ProductName')[['Quantity', 'TotalAmount']].sum().reset_index()
```

```python
product_sales_sorted = product_sales.sort_values(by='TotalAmount', ascending=False)
```

```python
top_10_products = product_sales_sorted.head(10)
print("--- Top 10 Products by Revenue ---")
print(top_10_products)
```

```python
category_sales = df.groupby('Category')['TotalAmount'].sum().reset_index()
```

```python
category_sales_sorted = category_sales.sort_values(by='TotalAmount', ascending=False)
print("\n--- Sales by Category ---")
print(category_sales_sorted)
```

### **Ôº≤ÔΩÖÔΩáÔΩâÔΩèÔΩéÔΩÅÔΩå ÔΩÅÔΩéÔΩÑ Ôº£ÔΩÅÔΩîÔΩÖÔΩáÔΩèÔΩíÔΩâÔΩÉÔΩÅÔΩå Ôº°ÔΩéÔΩÅÔΩåÔΩôÔΩìÔΩâÔΩì**

```python
regional_category_sales = df.groupby(['State', 'Category'])[['TotalAmount', 'Quantity']].sum().reset_index()
```

```python
state_sales = df.groupby('State')['TotalAmount'].sum().reset_index().sort_values(by='TotalAmount', ascending=False)
top_10_states = state_sales.head(10)
print("--- Top 10 States by Revenue ---")
print(top_10_states)
```

```python
payment_method_sales = df.groupby('PaymentMethod')['TotalAmount'].sum().reset_index()
print("\n--- Sales by Payment Method ---")
print(payment_method_sales)
```

```python
order_status_analysis = df.groupby('OrderStatus').agg({'TotalAmount': 'sum', 'OrderID': 'count'}).rename(columns={'OrderID': 'OrderCount'}).reset_index()
print("\n--- Order Status Analysis ---")
print(order_status_analysis)
```

### **Ôº∂ÔΩâÔΩìÔΩïÔΩÅÔΩåÔΩâÔΩöÔΩÅÔΩîÔΩâÔΩèÔΩéÔΩì**

```python
sns.set_theme(style='whitegrid')


fig, axes = plt.subplots(2, 2, figsize=(20, 15))


monthly_sales_display = monthly_sales.copy()
monthly_sales_display['YearMonthStr'] = monthly_sales_display['YearMonth'].astype(str)
sns.lineplot(ax=axes[0, 0], data=monthly_sales_display, x='YearMonthStr', y='TotalAmount', marker='o', color='teal')
axes[0, 0].set_title('Monthly Sales Trends Over Time', fontsize=16)
axes[0, 0].set_xlabel('Month', fontsize=12)
axes[0, 0].set_ylabel('Total Revenue', fontsize=12)
axes[0, 0].tick_params(axis='x', rotation=45)


sns.barplot(ax=axes[0, 1], data=top_10_products, x='TotalAmount', y='ProductName', hue='ProductName', palette='viridis', legend=False)
axes[0, 1].set_title('Top 10 Products by Revenue', fontsize=16)
axes[0, 1].set_xlabel('Total Revenue', fontsize=12)
axes[0, 1].set_ylabel('Product Name', fontsize=12)


sns.barplot(ax=axes[1, 0], data=top_10_states, x='State', y='TotalAmount', hue='State', palette='magma', legend=False)
axes[1, 0].set_title('Top 10 States by Revenue', fontsize=16)
axes[1, 0].set_xlabel('State', fontsize=12)
axes[1, 0].set_ylabel('Total Revenue', fontsize=12)


axes[1, 1].pie(category_sales_sorted['TotalAmount'], labels=category_sales_sorted['Category'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3'))
axes[1, 1].set_title('Revenue Distribution by Category', fontsize=16)


plt.tight_layout()
plt.show()
```
